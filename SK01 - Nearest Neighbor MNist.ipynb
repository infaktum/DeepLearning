# SK01 - Nearest Neighbor Klassifizierer (sklearn)

## Laden der Iris-Datenbank

```python
from sklearn.datasets import load_iris
iris = load_iris()

x = iris.data
y = iris.target

x_names = iris.feature_names
y_names = iris.target_names
```

## Laden der Iris-Abbildungen

```python
from PIL import Image
from IPython.display import display

images={0:Image.open('images/IrisSetosa.jpg','r'),
        1:Image.open('images/IrisVersicolor.jpg','r'),
        2:Image.open('images/IrisVirginica.jpg','r')}

def show_iris(iris):
    display(images[iris])

show_iris(1)
```

## Aufteilen der Daten in Trainings- und Test-Daten

```python
import numpy as np
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

## Training des Klassifizierers auf den Trainingsdaten

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors = 1)
knc.fit(x_train,y_train)
```

## Test des Klassifizierers

```python
knc.score(x_test,y_test)
```

```python
knc.predict(x_test) == y_test
```
